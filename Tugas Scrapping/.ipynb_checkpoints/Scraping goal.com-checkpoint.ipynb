# Tugas Scraping

# Scraping Goal.com
Pada python notebook ini akan dijelaskan mengenai scraping berita sepakbola www.goal.com. Scraping dimulai dari halaman awal, lalu akan menelusuri setiap link, dan akan mengambil berita beserta properties lainnya.

## Inisialisasi urllib.request

```python
import urllib.request

r = urllib.request.urlopen('https://www.liputan6.com/news').read()
print(type(r))
print(r[1:1000])
```

## Inisialisasi BeautifulSoup

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(r,"lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

## Dapatkan Semua Link

```python
links = soup.find_all('a')
print(len(links))

for link in links:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

Link yang benar-benar menuju ke berita adalah link yang mengandung '/berita/','/pertandingan/','/galeri/' dan benar-benar link bukan javascript maupun link '#', sehingga perlu difilter sebagai berikut:

```python
list_key_link = ['/read/']
berita_link = []
for i in list_key_link:
    berita_link = berita_link +[link for link in links if i in str(link) and 'javascript:void(0)' not in str(link) 
               and '#' not in str(link)]
print(len(berita_link))
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
```

## Dapatkan Isi Berita
Di sini akan diambil judul berita, isi berita, dan tanggal berita dipost/dibuat. Dari hasil inspeksi, judul ada pada elemen h1 dengan class read__title. Untuk isi berita ada pada elemen div class read__content. Sedangkan untuk tanggal ada pada elemen div read__time

```python
html = urllib.request.urlopen("https://www.liputan6.com/news/read/3678213/jokowi-minta-timnas-indonesia-u-19-berikan-yang-terbaik-saat-lawan-jepang").read()
soup = BeautifulSoup(html, "lxml")

judul = soup.find("h1","read-page--header--title entry-title").text.strip()
isi=''
#mengambil semua text dari div body
isi_berita = soup.find('div','read-page--content')
tanggal = soup.find("time","read-page--header--author__datetime updated").text[0:11]
print(tanggal)
print(judul)
for x in isi_berita.find_all('p'):
    isi=isi+x.text.strip()
isi
```

Menyimpan gambar dari berita, gambar ada pada div class photo

```python
gambar = soup.find("div","read-page--top-media").find("img","read-page--photo-gallery--item__picture-lazyload")
print(gambar['data-src'])

from PIL import Image
from io import BytesIO
import string

r = urllib.request.urlopen(gambar["data-src"]).read()
print(r)

i = Image.open(BytesIO(r))
exclude = set(string.punctuation)
judul = ''.join(str(ch) for ch in judul if ch not in exclude)

nama_file = "gambars/" + judul + ".png"
i.save(nama_file,'PNG')
```

## Fungsi Scraping
merupakan fungsi scraping yang utuh untuk scraping berita dari goal.com

```python
def getBerita(link):
    
    html = urllib.request.urlopen("https://www.liputan6.com/news").read()
    soup = BeautifulSoup(html, "lxml")
    judul = soup.find("h1","read-page--header--title entry-title").text.strip()
    isi=''
    isi_berita = soup.find('div','read-page--content')
    tanggal = soup.find("time","read-page--header--author__datetime updated").text[0:11]
    print(tanggal)
    print(judul)
  
    for x in isi_berita.find_all('p'):
        isi=isi+x.text.strip()     
    isi    
    #gambar
    gambar = soup.find("div","read-page--top-media").find("img","read-page--photo-gallery--item__picture-lazyload")
    print(gambar['data-src'])

    from PIL import Image
    from io import BytesIO
    import string

    r = urllib.request.urlopen(gambar["data-src"]).read()
    print(r)

    i = Image.open(BytesIO(r))
    exclude = set(string.punctuation)
    judul = ''.join(str(ch) for ch in judul if ch not in exclude)

    nama_file = "gambars/" + judul + ".png"
    i.save(nama_file,'PNG')
    
    return [tanggal,judul,isi]
```

## Dapatkan semua data dari link berita

```python
# getBerita('https://ekonomi.kompas.com/read/2018/10/24/204900026/pertamina-ingin-beli-minyak-dengan-mata-uang-selain-dollar-as')
berita = []
for link in berita_link:
    print('%s: %s'%(link.text.strip(),link["href"]))
    try:
        berita.append(getBerita(link["href"]))
    except:
        pass
```

```python
import pandas as pd
df =pd.DataFrame(berita, columns=['tanggal','judul','isi'])
df
```

## Simpan ke dalam CSV

```python
df.to_csv('data_berita_bola.csv',index=False, encoding='utf-8')
```

## Simpan ke dalam XLSX

```python
import pandas as pd
data = pd.DataFrame(berita)
writer = pd.ExcelWriter('data_berita.xlsx')
data.to_excel(writer,'Sheet1',index=False)
writer.save()
```
